## 数据类型

### 1. str – 字符串

````python
# 注意 单引号和双引号无明显区别

## 优先使用单引号：用于表示普通的、简短的字符串（如字典的键、变量值）。
## 优先使用双引号：用于自然语言（如用户界面显示的句子）或包含单引号的内容。

########## 定义
s1 = 'hello'          # 单引号
s2 = "world"          # 双引号
s3 = '''多行
字符串'''              # 三引号支持换行
s4 = "It's OK"        # 双引号内含单引号无冲突

########## 常用操作

# 拼接
s = 'Hello' + ' ' + 'Python'  # 'Hello Python'

# 重复
s = 'Ha' * 3                  # 'HaHaHa'

# 索引和切片
s = 'abcdef'
print(s[0])      # 'a'
print(s[-1])     # 'f'
print(s[1:4])    # 'bcd'（左闭右开）
print(s[::-1])   # 'fedcba'（反转）

# 长度
len('hello')      # 5

# 成员判断
'a' in 'abc'      # True
'x' not in 'abc'  # True

# 大小写转换
'Hello'.lower()   # 'hello'
'Hello'.upper()   # 'HELLO'
'hello'.capitalize()  # 'Hello'
'title case'.title()  # 'Title Case'

# 去除空白
'  text  '.strip()   # 'text'
'  text'.lstrip()    # 'text'
'text  '.rstrip()    # 'text'

# 分割与合并
'one,two,three'.split(',')  # ['one', 'two', 'three']
'-'.join(['a','b','c'])     # 'a-b-c'

# 替换
'hello world'.replace('world', 'Python')  # 'hello Python'

# 查找
'abcabc'.find('b')   # 1 (首次出现索引)
'abcabc'.count('a')  # 2

# 格式化（三种常见方式）
name = 'Alice'
age = 30
# % 格式化
'My name is %s, age %d' % (name, age)
# format 方法
'My name is {}, age {}'.format(name, age)
# f-string（Python 3.6+）
f'My name is {name}, age {age}'

# 判断数字/字母等
'123'.isdigit()   # True
'abc'.isalpha()   # True
'a1'.isalnum()    # True (字母数字)
'   '.isspace()   # True

########## 类型转换（转为字符串） 
str(123)          # '123'
str(3.14)         # '3.14'
str(True)         # 'True'
str(None)         # 'None'
````

### 2. int – 整数

````python
## 整数表示正整数、负整数和零，无精度限制（Python 3 支持大整数）

########## 定义与进制
a = 42            # 十进制
b = 0b1010        # 二进制 (10)  binary
c = 0o12          # 八进制 (10) octal
d = 0xA           # 十六进制 (10) hex

########## 常用操作
# 四则运算
10 + 3    # 13
10 - 3    # 7
10 * 3    # 30
10 / 3    # 3.333... (返回 float)
10 // 3   # 3 (整除)
10 % 3    # 1 (取余)
10 ** 3   # 1000 (幂)

# 位运算（仅整数）
5 & 3     # 1 (AND)     按位与:	全 1 为 1（有 0 则 0）;         0101 & 0011  =   0001  =  1
5 | 3     # 7 (OR)      按位或:	有 1 为 1（全 0 则 0）;         0101 | 0011  =   0111  =  1+2+4 = 7
5 ^ 3     # 6 (XOR)     按位异或:	相同为 0，不同为 1;         0101 ^ 0011  =   0110  =   2+4 = 6
~5        # -6 (取反)   按位取反: 通用公式：~x = -(x + 1)；	0 变 1，1 变 0（单目运算）   0000 0101 取反  1111 1010 （计算机存储这段数字，而计算机存储的都是补码，所以这是一段补码），符号为不动，再次取反+1即可获得值
5 << 1    # 10 (左移)   左移:	整体左移，右边补 0，左边溢出丢弃。  0101 ： 0000 0101 - 0000 1010
5 >> 1    # 2 (右移)    右移:	整体右移，左边补符号位（正数补0），右边丢弃。  010 ： 0000 0101 -  0000 0010


# 绝对值、符号
abs(-5)   # 5
pow(2, 3) # 8 (2的3次方)
divmod(10, 3)  # (3, 1) 商和余数

# 进制转换
bin(10)   # '0b1010'
oct(10)   # '0o12'
hex(10)   # '0xa'
int('1010', 2)  # 10 (将二进制字符串转为整数)

########## 类型转换（转为整数） 
int('123')          # 123
int('   -45  ')     # -45 (自动去除空白)
# int('3.14')         # 报错 (不能直接转浮点字符串)
int(3.14)           # 3 (截断小数)
int(True)           # 1
int(False)          # 0
int('0b1010', 2)    # 10 (指定进制)
int('A', 16)        # 10

########## 注意事项
# 除法 / 总是返回 float，若需要整数除法用 //。
# 大整数运算自动支持，无需担心溢出。
````

### 3. float – 浮点数

````python
# 浮点数表示实数，通常实现为双精度（64位），有精度限制。

######## 定义

pi = 3.14159
x = 1.0
y = 2.5e-3   # 0.0025 (科学计数法)
z = .5       # 0.5 (省略前导0)

######## 常用操作

# 基本运算（与 int 类似，但结果通常为 float）
3.0 + 2.0    # 5.0
3.0 / 2.0    # 1.5
3.0 // 2.0   # 1.0 (整除，返回 float)
3.0 ** 2.0   # 9.0

# 特殊值
float('inf')     # 正无穷
float('-inf')    # 负无穷
float('nan')     # 非数 (Not a Number)

# 判断
import math
math.isinf(x)    # 是否无穷
math.isnan(x)    # 是否 NaN

# 四舍五入
round(3.14159, 2)  # 3.14
round(2.675, 2)    # 2.67 (注意浮点精度，并非预期 2.68)

# 取整（向零截断）
int(3.9)    # 3
int(-3.9)   # -3

# 绝对值
abs(-3.14)  # 3.14

#######  类型转换（转为浮点数）

float('3.14')      # 3.14
float('1.2e-3')    # 0.0012
float(42)          # 42.0
float(True)        # 1.0
float(False)       # 0.0
float('inf')       # inf


######## 注意事项
## 浮点运算存在精度误差（如 0.1 + 0.2 != 0.3），比较时建议使用 abs(a-b) < 1e-9。
## 避免用 == 直接比较浮点数。
````

### 4. bool – 布尔值

````python
# 布尔值只有 True 和 False，是 int 的子类（True 等价于 1，False 等价于 0）。

####### 定义与判断
flag = True
flag = False

###### 布尔运算
# 逻辑运算
True and False   # False
True or False    # True
not True         # False

# 比较运算（返回 bool）
3 > 2            # True
3 == 2           # False
3 != 2           # True
'abc' == 'abc'   # True
[1,2] == [1,2]   # True

# 身份运算（is 比较对象 ID）
a = None
a is None        # True
True is True     # True（小整数和布尔值有缓存）

######## 类型转换（转为布尔值）
bool(42)          # True (非零数值为 True)
bool(3.14)        # True
bool('hello')     # True (非空字符串为 True)
bool([1,2])       # True

bool(0)           # False
bool(0.0)         # False
bool('')          # False
bool([])          # False (空容器为 False)
bool(None)        # False

######### 常见用法

# 条件语句
if flag:
    print("True")

# 使用 bool 作为整数
sum([True, False, True])  # 2 (True 当作 1)

# 使用 any/all 判断可迭代对象  
any([False, False, True])  # True
all([True, True, False])   # False  判断一个可迭代对象（iterable）里的所有元素，是否全部为真（Truthy）
````

### 5. None – 空值

````python
# None 是 NoneType 类型的唯一实例，表示“没有值”或“空”。常用于函数默认返回值、占位符等。

########## 定义与比较
x = None
y = None
x is y    # True（单例对象）

# 不要用 == 比较，推荐使用 is
if x is None:
    print("x is None")


######### 类型转换

str(None)    # 'None'
# int(None)    # 报错 (无法转换)
bool(None)   # False

######## 常见使用场景

# 函数默认返回 None
def foo():
    return  # 无 return 语句，默认返回 None
print(foo())  # None

# 作为参数默认值
def process(data=None):
    if data is None:
        data = []
    # ...

# 作为占位符，表示缺失
config = {'host': 'localhost', 'port': None}  # port 未指定

######### 注意事项
## None 不等于空字符串、0 或 False，它们是不同的对象。
## 函数若没有显式 return，会返回 None。
````

````python
# 特点：有序、可变（增删改）、元素可重复、可存放任意类型。

######### 定义与基本操作
# 定义
lst = [1, 2, 3, 'hello', True]
empty = []

# 索引与切片（同字符串）
print(lst[0])       # 1
print(lst[-1])      # True
print(lst[1:4])     # [2, 3, 'hello']
print(lst[::-1])    # 反转 [True, 'hello', 3, 2, 1]

# 长度
len(lst)            # 5

######## 增删改
nums = [1, 2, 3]

# 增加
nums.append(4)          # [1, 2, 3, 4]  末尾追加
nums.insert(1, 99)      # [1, 99, 2, 3, 4]  指定位置插入
nums.extend([5, 6])     # [1, 99, 2, 3, 4, 5, 6]  合并列表

# 删除
nums.pop()              # 弹出末尾元素 6，并返回
nums.pop(1)             # 弹出索引1的元素 99
nums.remove(2)          # 删除第一个值为2的元素
del nums[0]             # 删除索引0的元素
# nums.clear()          # 清空所有

# 修改
nums[0] = 100           # 直接赋值修改

######### 查找与排序

nums = [3, 1, 4, 1, 5, 9, 2]

# 查找
nums.index(4)           # 2 (首次出现索引)
nums.count(1)           # 2 (统计出现次数)
5 in nums               # True
10 not in nums          # True

# 排序（原地修改）
nums.sort()             # [1, 1, 2, 3, 4, 5, 9] (升序)
nums.sort(reverse=True) # [9, 5, 4, 3, 2, 1, 1] (降序)
nums.reverse()          # 反转顺序

# 不改变原列表的排序
sorted(nums)            # 返回新列表

########## 列表推导式（高频用法）
# 生成 0~9 的平方列表
squares = [x**2 for x in range(10)]          # [0, 1, 4, ..., 81]

# 带条件过滤（取偶数）
evens = [x for x in range(10) if x % 2 == 0] # [0, 2, 4, 6, 8]

# 嵌套循环（展平矩阵）
matrix = [[1,2], [3,4]]
flat = [num for row in matrix for num in row] # [1, 2, 3, 4]

test1 = [x**2 for x in range(10)]
test2 = [x for x in range(10) if x % 2 == 0 ]
print("test1:",test1)
print("test2:",test2)

############# 复制（重要）
a = [1, 2, 3]
b = a            # 引用传递，修改b会影响a
c = a.copy()     # 浅拷贝（或 a[:]）
import copy
d = copy.deepcopy(a)  # 深拷贝（嵌套列表时用）: 会递归地把所有内部元素都复制一份新的，彻底切断关系.而不是“把嵌套结构拍扁”。

b[0] = 4  # 会修改a的数据
c[0] = 5  # 不会修改a的数据
d[0] = 6
print("a:",a)
print("b:",b)
print("c:",c)
print("d:",d)

test3 = [[1,2,3], [3,4]]
test3_a = [num for row in test3 for num in row]

test3_b = test3.copy()
test3_b[1].append(5)

test3_c = copy.deepcopy(test3)
test3_c[1].append(10)
test3_d = [num for row in test3_c for num in row]
print("test3:",test3)
print("test3_b:",test3_b)
print("test3_c:",test3_c)

#### 注意
## 浅拷贝：保护顶层容器，但不保护内部子对象。修改内部可变元素会影响原始数据。
## 深拷贝：保护所有层级，修改拷贝绝不会动原始数据（这才是真正的“值传递”效果）。
#####  深浅拷贝总结
## 非嵌套列表：浅拷贝不会改变原始数据，深拷贝也不会；嵌套列表：浅拷贝会改变原始数据，深拷贝不会

````

### 2. tuple – 元组

````python
# 特点：有序、不可变（创建后不能增删改）、元素可重复、可作为字典的键。

######### 定义与访问
# 定义（括号可省略）
t1 = (1, 2, 3)
t2 = 4, 5, 6          # 逗号是关键
t3 = (7,)             # 单元素元组必须带逗号，否则会被识别为 int

# 访问（同列表）
t1[0]      # 1
t1[-1]     # 3
t1[1:3]    # (2, 3)

# 遍历输出
t = ('a', 'b', 'c', 'd')

for item in t:
    print(item)
# 输出: a b c d (每行一个)
t = ('a', 'b', 'c', 'd')

# 需要索引时，使用:enumerate
for index, value in enumerate(t):
    print(f"索引 {index} 对应的值是 {value}")
# 输出: 索引 0 对应的值是 a ...

########### 不可变性（注意陷阱）
t = (1, 2, [3, 4])
# t[0] = 100  # 报错：不能修改元组元素
t[2][0] = 99  # 可以！因为元组存储的是列表的引用，列表本身可变
print(t)      # (1, 2, [99, 4])
# 案例
test1 = (1,2,4,[2,3],{5,6})
## 列表可以直接改
test1[3][1] = 55
## 集合
# 先移除 6，再添加 55
test1[4].remove(5)
test1[4].add(66)
print(test1)

########## 解包（Unpacking）——超级实用
point = (10, 20)
x, y = point          # x=10, y=20

# 带 * 号收集剩余元素
a, *rest = (1, 2, 3, 4)   # a=1, rest=[2, 3, 4]
*head, tail = (1, 2, 3)   # head=[1, 2], tail=3

# 交换变量（本质是元组解包）
a, b = b, a

########### 方法与转换
t = (1, 2, 2, 3)
t.count(2)     # 2
t.index(2)     # 1

# 转列表 / 字符串
list(t)        # [1, 2, 2, 3]
tuple([1,2])   # (1, 2)


````

### 3. dict – 字典 (键值对)

````python
# 特点：无序（Python 3.7+ 按插入顺序）、可变、键（Key）唯一且必须可哈希（不可变类型，如字符串、数字、元组）。

######### 定义与访问（安全取值）
# 定义
d1 = {'name': 'Alice', 'age': 30}
d2 = dict(name='Bob', age=25)   # 关键字形式


# 访问（推荐用 get，避免报错）
d1['name']          # 'Alice'
d1.get('age')       # 30
d1.get('gender', 'Unknown')  # 'Unknown' (不存在返回默认值)

# 获取所有键/值/项  返回的都是：视图对象；当字典发生变化时，这个视图对象会自动同步最新的键，无需重新生成。
d1.keys()    # dict_keys(['name', 'age'])
d1.values()  # dict_values(['Alice', 30])
d1.items()   # dict_items([('name', 'Alice'), ('age', 30)])

###### 增删改
d = {'a': 1, 'b': 2}

# 修改/新增
d['c'] = 3          # 新增键 'c'
d['a'] = 100        # 修改键 'a' 的值
d.update({'b': 200, 'd': 4})  # 批量更新/新增


# 删除
d.pop('a')          # 删除键 'a' 并返回值
d.popitem()         # 弹出最后插入的一个键值对 (Python 3.7+)
del d['b']          # 删除键 'b'
# d.clear()         # 清空

###### 字典推导式与合并
# 反转键值
original = {'a': 1, 'b': 2}
reversed_dict = {v: k for k, v in original.items()}  # {1: 'a', 2: 'b'}

# 合并（Python 3.9+）
d1 = {'x': 1, 'y': 2}
d2 = {'y': 3, 'z': 4}
merged = d1 | d2                    # {'x': 1, 'y': 3, 'z': 4}
# 或使用 d1.update(d2) 原地修改

########## 遍历（迭代）
d = {'name': 'Tom', 'age': 18}
for key in d:              # 遍历键
    print(key)
for value in d.values():   # 遍历值
    print(value)
for k, v in d.items():     # 遍历键值对（最常用）
    print(f"{k}: {v}")
````

### 4. set – 集合 (单独的元素)

````python
# 特点：无序、可变、元素唯一（自动去重）、元素必须可哈希。

###### 定义与去重
# 定义（注意：空集合必须用 set()，{} 是空字典）
s1 = {1, 2, 3, 3, 2}    # {1, 2, 3}  自动去重
s2 = set([1, 1, 2, 3])  # {1, 2, 3}  从列表创建

# 常用场景：列表去重
nums = [1, 2, 2, 3, 3, 3]
unique = list(set(nums))  # [1, 2, 3] (顺序可能变化)

######### 增删
s = {1, 2, 3}

# 增加
s.add(4)          # {1, 2, 3, 4}

# 删除（注意区别）
s.remove(4)       # 删除元素4，若不存在则报错 KeyError
s.discard(4)      # 删除元素4，若不存在则**不报错**（安全）
s.pop()           # 随机弹出一个元素
# s.clear()       # 清空

######## 集合运算（数学逻辑）——核心优势
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# 并集
A | B              # {1, 2, 3, 4, 5, 6}  或 A.union(B)

# 交集
A & B              # {3, 4}  或 A.intersection(B)

# 差集（属于A不属于B）
A - B              # {1, 2}  或 A.difference(B)

# 对称差集（不同时属于A和B的元素）
A ^ B              # {1, 2, 5, 6}  或 A.symmetric_difference(B)

# 子集/超集判断
{1, 2}.issubset(A)  # True
A.issuperset({1})   # True
A.isdisjoint(B)     # False (是否有交集)

########## 集合推导式
# 取列表中偶数的平方集合
lst = [1, 2, 2, 3, 4, 4]
even_squares = {x**2 for x in lst if x % 2 == 0}  # {16, 4} (自动去重)
````

## 控制流

### 1. if/elif/else – 条件判断

### 2. for – 迭代循环

````python
# 用于遍历任何可迭代对象（列表、元组、字典、字符串、range 等）

######## 遍历列表、元组、字符串
# 遍历列表
fruits = ['apple', 'banana', 'orange']
for fruit in fruits:
    print(fruit)   # 依次打印 apple, banana, orange

# 遍历字符串
for ch in 'hello':
    print(ch)      # h e l l o

######### 使用 range() 进行次数循环
# range(stop): 0 ~ 4
for i in range(5):
    print(i)       # 0, 1, 2, 3, 4

# range(start, stop, step): 1 ~ 10 的奇数
for i in range(1, 10, 2):
    print(i)       # 1, 3, 5, 7, 9

########### 遍历字典（最常用方式）
user = {'name': 'Tom', 'age': 20, 'city': 'Beijing'}

# 遍历键值对（最推荐）
for key, value in user.items():
    print(f'{key}: {value}')

# 仅遍历键
for key in user.keys():
    print(key)

# 仅遍历值
for value in user.values():
    print(value)

########## 同时获取索引和元素（enumerate）
fruits = ['apple', 'banana', 'orange']
for index, fruit in enumerate(fruits):
    print(f'索引 {index} 是 {fruit}')
# 输出: 索引 0 是 apple ...

######### 并行遍历多个列表（zip）
names = ['Alice', 'Bob', 'Charlie']
scores = [95, 87, 92]

for name, score in zip(names, scores):
    print(f'{name} 得分 {score}')


######### for 与 else 搭配（有 break 时不执行 else）
nums = [1, 3, 5, 7]
for n in nums:
    if n % 2 == 0:
        print('找到偶数')
        break
else:
    print('没有找到偶数')  # 会执行这一行，因为循环没有被 break 中断

########### break 和 continue
# break: 终止整个循环
for i in range(5):
    if i == 3:
        break
    print(i)      # 输出 0, 1, 2

# continue: 跳过本次循环
for i in range(5):
    if i % 2 == 0:
        continue
    print(i)      # 输出 1, 3
````

### 2. *args —— 可变位置参数 

#### 核心是*的使用
* 在 Python 中扮演“打包”和“解包”两种角色
* 打包成元组：✅ 正确（函数定义时）。
* 解包列表、元组、集合：✅ 正确（函数调用时）。
* 解包字典：⚠️ 单星号 * 能解包字典，但只解包出“键（keys）”；如果想解包“键值对”作为参数，必须用 双星号 **。

````python
# *args 对传入的数据类型没有限制，它只是把传进来的“一堆东西”收集到一个元组容器里
# 用于接收任意数量的位置参数，打包成一个元组（tuple）。

########## 基本用法：求和或打印
# 计算任意个数的数字之和
def sum_all(*args):
    print(f"接收到: {args}")   # 类型为 tuple
    return sum(args)

def args_all(*args):
    return args

print(sum_all(1, 2, 3))      # 输出: 接收到: (1, 2, 3)  -> 6
print(args_all(10, 20,[4,6]))       # 输出: 接收到: (10, 20,[4,6]) ->  (10, 20, [4, 6])
print(args_all("hello",666,[23,24],{44,68}))  # ('hello', 666, [23, 24], {44, 68})

############ 与位置参数混用
def introduce(greeting, *names):
    for name in names:
        print(f"{greeting}, {name}")

introduce("Hello", "Tom", "Jerry", "Spike")
# 输出:
# Hello, Tom
# Hello, Jerry
# Hello, Spike

########### 反向操作：解包列表/元组传入
def add(a, b, c):
    return a + b + c

nums = [1, 2, 3]
print(add(*nums))   # 6  (把列表拆成三个位置参数)

nums2 = (4,5,6)
print(add(*nums2))

nums3 = {7,8,9}
print(add(*nums3))
````

### 3. **kwargs —— 可变关键字参数

````python
# 用于接收任意数量的关键字参数，打包成一个字典（dict）

########## 基本用法：处理配置或打印键值对
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="Beijing")
# 输出:
# name: Alice
# age: 25
# city: Beijing

########### 用于参数转发（常见于装饰器或封装函数）
def log_request(method, url, **extra):
    print(f"请求方法: {method}, URL: {url}")
    if extra:
        print(f"额外参数: {extra}")

log_request("POST", "/api", data="json", timeout=5)
# 输出:
# 请求方法: POST, URL: /api
# 额外参数: {'data': 'json', 'timeout': 5}

############### 反向操作：解包字典传入
def user_profile(name, age, city):
    print(f"{name}, {age}岁, 来自{city}")

data = {"name": "Bob", "age": 30, "city": "Shanghai"}
user_profile(**data)   # 把字典拆成 key=value 传入
````

### 🚀 *和**核心对比表

| 场景                   | `*`（单星号）                                  | `**`（双星号）                                    |
| :--------------------- | :--------------------------------------------- | :------------------------------------------------ |
| **打包（定义函数时）** | `def func(*args):`                             | `def func(**kwargs):`                             |
| **打包后的类型**       | **元组 (`tuple`)**                             | **字典 (`dict`)**                                 |
| **打包接收的数据**     | 任意数量的**位置参数**（如 `1, 2, 3`）         | 任意数量的**关键字参数**（如 `a=1, b=2`）         |
| **打包的限制**         | ✅ **无限制**，可以接收任何类型的值             | ⚠️ **有限制**，键（Key）必须是字符串且为合法标识符 |
| **解包（调用函数时）** | `func(*[1,2,3])`                               | `func(**{'a':1, 'b':2})`                          |
| **解包的数据来源**     | 任何**可迭代对象**（list, tuple, set, str...） | **字典**（或映射类型）                            |
| **解包后的传递方式**   | 拆解为**位置参数**（按顺序对应）               | 拆解为**关键字参数**（按名字对应）                |
| **解包的核心要求**     | 元素个数必须与函数参数个数匹配                 | 字典的键必须与函数参数名匹配                      |


### 4. 参数组合使用（正确的顺序）

````python
# 在同一个函数中，四类参数必须按以下固定顺序排列：
######### 位置参数 > *args > 默认参数 > **kwargs
def complex_func(a, b, *args, c=10, d=20, **kwargs):
    print(f"位置参数: a={a}, b={b}")
    print(f"可变位置: {args}")
    print(f"默认参数: c={c}, d={d}")
    print(f"可变关键字: {kwargs}")

complex_func(1, 2, 3, 4, c=100, name="test")
# 输出:
# 位置参数: a=1, b=2
# 可变位置: (3, 4)
# 默认参数: c=100, d=20
# 可变关键字: {'name': 'test'}

## 记忆口诀：a（位置）, *args（星号元组）, c=10（默认）, **kwargs（双星字典）。
````

### 5. lambda —— 匿名函数  
* lambda 就是 Python 为你准备的“一次性便签纸”
* 简洁、一次性使用、没有名字
* 它的本质就像一个“即用即抛”的小工具，专门用来处理那些逻辑非常简单、不值得专门用 def 去定义一个完整函数的场景

````python
# 用于定义单行表达式的简单函数，无需 def 和 return，常用于需要函数对象的场合（如排序、映射）

######### 基本语法
# 普通函数
def add(x, y):
    return x + y

# Lambda 写法
add_lambda = lambda x, y: x + y

print(add_lambda(3, 5))   # 8

##### 高频场景 1：与 sorted() 配合自定义排序
students = [("Alice", 25), ("Bob", 20), ("Charlie", 23)]

# 按年龄排序（元组第二个元素） .sort：仅限列表
students.sort(key=lambda stu: stu[1])
print(students)  # [('Bob', 20), ('Charlie', 23), ('Alice', 25)]

# 按字典中的某个值排序  sorted:任何可迭代对象（如列表、元组、集合、字典、生成器等）
users = [{"name": "Tom", "age": 18}, {"name": "Jerry", "age": 22}]
sorted_users = sorted(users, key=lambda u: u["age"])

######### 高频场景 2：与 map() / filter() 搭配
nums = [1, 2, 3, 4, 5]

# 计算平方（map 映射）
squares = list(map(lambda x: x**2, nums))  # [1, 4, 9, 16, 25]

# 过滤偶数（filter 筛选）
evens = list(filter(lambda x: x % 2 == 0, nums))  # [2, 4]

########## 高频场景 3：在 GUI/回调中作为临时处理函数
# 假设按钮点击事件
button = Button(text="点击", command=lambda: print("按钮被点击了！"))

######### ⚠️ Lambda 限制与陷阱
## 只能包含单个表达式，不能包含语句（如 return、while、print 返回 None 但可使用）。
## 避免在循环中创建 lambda 引用循环变量（经典坑）：
# 错误 ❌ (所有 lambda 都返回 9，因为 i 是循环结束后的值)
funcs = [lambda: i for i in range(5)]
print(funcs[0]())  # 4

# 正确 ✅ (默认参数捕获当前值)
funcs = [lambda x=i: x for i in range(5)]
print(funcs[0]())  # 0
````

### 6. 综合实战示例

````python
# 一个结合上述所有特性的日志记录函数

import datetime

def log(level, message, *tags, prefix="[APP]", **extra):
    """
    用法: log("INFO", "启动成功", "system", "boot", user="admin")
    """
    time_str = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    parts = [prefix, time_str, level, message]
    if tags:
        parts.append("标签: " + ", ".join(tags))
    if extra:
        parts.append("附加: " + ", ".join(f"{k}={v}" for k, v in extra.items()))
    print(" ".join(parts))

# 调用
log("ERROR", "连接超时", "network", "retry", user="test", attempt=3)
# 输出示例: [APP] 2023-10-01 12:00:00 ERROR 连接超时 标签: network, retry 附加: user=test, attempt=3
````

## try/except/finally
* Python 的异常处理机制 try/except/else/finally 是编写健壮代码的核心
* 它不仅能防止程序因报错而崩溃，还能优雅地处理错误、释放系统资源。

### 1. 基础结构：try / except

````python
try:
    num = int(input("请输入数字："))
    result = 10 / num
    print(f"结果是：{result}")
except ValueError:
    print("错误：请输入有效的整数！")  # 输入字母时会触发
except ZeroDivisionError:
    print("错误：除数不能为 0！")      # 输入 0 时会触发
````

### 2. 捕获异常实例（获取错误信息）

````python
try:
    with open("data.txt", "r") as f:
        content = f.read()
except FileNotFoundError as e:
    print(f"文件未找到，系统提示：{e}")  
    # 输出：文件未找到，系统提示：[Errno 2] No such file or directory: 'data.txt'
````

### 3. else 子句（没有异常时执行）
* else 里的代码仅在 try 块没有发生任何异常时运行。这能让 try 块只放“危险”代码，逻辑更清晰。

````python
try:
    num = int(input("输入被除数："))
    divisor = int(input("输入除数："))
    result = num / divisor
except ValueError:
    print("输入必须为数字")
except ZeroDivisionError:
    print("除数不能为0")
else:
    # 只有计算成功时才执行
    print(f"计算成功，结果是：{result}")
````

### 4. finally 子句（无论如何都执行）
* finally 里的代码无论是否发生异常，都会在最后执行，专门用于释放资源（关闭文件、数据库连接、网络套接字等）
* 💡 提示：在实际开发中，操作文件、数据库连接通常推荐使用 with 语句（上下文管理器），它会自动调用 close()，但 finally 在处理自定义锁、计数器等场景下依然不可或缺。

````python
# 手动管理文件资源（经典场景）
f = None
try:
    f = open("test.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("文件不存在")
finally:
    # 确保文件被关闭，避免内存泄漏
    if f is not None:
        f.close()
        print("文件已安全关闭")
````

````python
try:
    risky_function()
except Exception as e:
    print(f"发生未知错误：{e}")  # 记录日志，避免程序崩溃
    # 注意：这里不要吞掉异常，至少打印或记录日志
````

### 6. 手动抛出异常（raise）
* 在业务逻辑中，如果条件不满足，可以主动抛出异常，将错误向上层传递。

````python
import json

def load_config(filepath):
    config = None
    try:
        # 尝试打开并解析 JSON
        f = open(filepath, "r")
        raw_data = f.read()
        config = json.loads(raw_data)
    except FileNotFoundError:
        print(f"错误：配置文件 {filepath} 不存在")
    except json.JSONDecodeError as e:
        print(f"错误：JSON 格式解析失败 - {e}")
    else:
        # 没有异常时，标记加载成功
        print("配置文件加载成功！")
        return config
    finally:
        # 无论成败，确保文件对象被关闭（预防资源泄漏）
        # 注意：这里要处理 f 可能未定义的情况
        if 'f' in locals() and not f.closed:
            f.close()
            print("文件资源已释放")

    return config  # 出错返回 None

# 测试
load_config("setting.json")
````

### 1. 类定义、__init__ 与实例方法
__init__ 是构造函数，用于初始化实例属性。实例方法的第一个参数必须是 self，指向实例本身

````python
class Student:
    # 类属性（所有实例共享）
    school = "北京大学"

    # 构造方法：初始化实例
    def __init__(self, name, age, score):
        self.name = name          # 实例属性
        self.age = age
        self.score = score
        # 私有属性（约定：双下划线开头，实际会被伪装）
        self.__id = "S001"

    # 实例方法（最常用）
    def introduce(self):
        return f"我叫 {self.name}，{self.age} 岁"

    # 实例方法（带业务逻辑）
    def get_grade(self):
        if self.score >= 90:
            return "A"
        elif self.score >= 80:
            return "B"
        else:
            return "C"

# 使用
s1 = Student("张三", 20, 95)
print(s1.introduce())       # 我叫 张三，20 岁
print(s1.get_grade())       # A
print(s1.school)            # 北京大学（访问类属性）
````

### 2. 继承（Inheritance）
子类可以继承父类的属性和方法，并可以重写（Override）或扩展（通过 super()）父类逻辑

⚠️ 重要提醒：子类重写 __init__ 时，必须用 super() 调用父类的 __init__，否则父类属性不会被初始化，导致运行时报错。

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        return f"{self.name} 在说话"

# 子类继承 Person
class Teacher(Person):
    def __init__(self, name, age, subject):
        # 调用父类构造方法（关键步骤） 执行父类 __init__ 里的所有逻辑。
        super().__init__(name, age)
        self.subject = subject

    # 重写父类方法
    def speak(self):
        # 复用父类逻辑并扩展
        base = super().speak()
        return f"{base}，他是教 {self.subject} 的老师"

# 使用
t = Teacher("李老师", 35, "数学")
print(t.speak())  
# 输出：李老师 在说话，他是教 数学 的老师
````

### 3. 魔法方法（__str__ 与 __repr__）
虽然不是必须，但强烈建议掌握，用于调试和打印对象信息。

````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    # 面向用户（print 时调用）
    def __str__(self):
        return f"《{self.title}》- {self.author}"

    # 面向开发者（交互式环境直接输入对象时显示） 如果没有__str__ 则执行__repr__
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}')"

b = Book("三体", "刘慈欣")
print(b)   # 调用 __str__：《三体》- 刘慈欣
# 在交互式环境直接输入 b，会显示 Book('三体', '刘慈欣')
````

### 1. dataclass（数据类，Python 3.7+）
dataclass 主要用于存储数据的类，它会自动帮我们生成 __init__、__repr__、__eq__ 等样板代码，让代码非常简洁。

需要手动加的魔术方法：
| 分类           | 魔术方法                           | 触发时机          | 典型应用                                    |
| :------------- | :--------------------------------- | :---------------- | :------------------------------------------ |
| **算术运算符** | `__add__(self, other)`             | `obj1 + obj2`     | 实现向量加法（`Vector(1,2) + Vector(3,4)`） |
| **比较运算符** | `__lt__(self, other)`              | `obj1 < obj2`     | 实现排序（`sorted(books)；只要定义了 __lt__，sorted()、min()、max() 都能自动工作。`）                 |
| **容器模拟**   | `__getitem__(self, key)`           | `obj[key]`        | 让你的对象像列表/字典一样取；还可以迭代（for item in ml: 自动生效）值               |
| **上下文管理** | `__enter__`, `__exit__`            | `with obj as ...` | 自定义资源管理器（如数据库连接）            |
| **描述符协议** | `__get__`, `__set__`, `__delete__` | 属性访问          | `@property` 的底层实现                      |


适用场景：主要用于内部数据传递、配置对象、DTO（数据传输对象）。
如果类中包含复杂业务逻辑，不建议用 dataclass，用普通类更合适。

````python
from dataclasses import dataclass, field
from typing import List

@dataclass
class UserData:
    username: str
    age: int
    # 带默认值（注意：有默认值的要放在后面）
    is_active: bool = True
    # 处理可变默认值（如列表），必须用 field(default_factory=list)  否则会导致实例共享
    tags: List[str] = field(default_factory=list)  # compare=False ： 会放弃这个字段的比较

# 使用（无需手动写 __init__）
user = UserData("alice", 25, tags=["admin", "vip"])
print(user)          # UserData(username='alice', age=25, is_active=True, tags=['admin', 'vip'])

# 自动支持相等比较
user2 = UserData("alice", 25)
print(user == user2)  # True (因为 __eq__ 自动生成)
````

### 2. Pydantic BaseModel（Web 框架的标配）
这是 FastAPI、Django Ninja、LangChain 等框架的核心数据验证组件。它与 dataclass 最大的不同是强校验 + 自动类型转换 + 序列化（支持 JSON）。

核心用途：

请求体校验（如 FastAPI 的 Body 参数）。

环境变量配置管理（继承 BaseSettings）。

数据序列化/反序列化（ORM 对象转 JSON）

````python
from pydantic import BaseModel, EmailStr, validator, Field,field_validator
from typing import Optional

class UserSignup(BaseModel):
    # 类型注解：EmailStr 会校验邮箱格式
    email: EmailStr
    password: str
    # Optional 表示可选，Field 用于添加校验规则
    age: Optional[int] = Field(None, ge=0, le=200)
    name: str = Field(..., min_length=2, max_length=20)  # ... 表示必填

    # 自定义校验器（例如禁止密码为 "123456"）
    @field_validator('password')
    def password_strong(cls, v):
        if v == "123456":
            raise ValueError("密码太弱")
        return v

# 1. 数据验证与自动转换
try:
    # age 传了字符串 "18"，Pydantic 会自动转为 int
    data = UserSignup(
        email="test@example.com",
        password="secure123",
        age="18",       # 注意：自动转 int，太方便了！
        name="Bob"
    )
    print(data.age)     # 18 (int 类型)
    print(data.model_dump_json())  # 转 JSON 字符串（FastAPI 响应时自动调用）
except ValueError as e:
    print(e)

# 2. 校验失败示例
try:
    invalid = UserSignup(
        email="not_email",
        password="123456",
        name="B"
    )
except Exception as e:
    print(e.json())  # 打印详细的 JSON 格式报错（框架用它返回 422 错误）
````

| 场景                                 | 推荐方案                 | 理由                                          |
| :----------------------------------- | :----------------------- | :-------------------------------------------- |
| 复杂业务逻辑（如交易引擎、游戏角色） | **普通类 + `__init__`**  | 灵活、可控，便于维护复杂的内部状态和方法。    |

3. **Pydantic V2 注意**：`@validator` 在 V2 中改为 `@field_validator`，语法略有不同（但逻辑不变），建议查对应版本文档。

## 4. asyncio —— 异步编程入门
* 异步编程让单线程能并发执行 I/O 密集型任务（如网络请求、数据库查询）。基础三要素：async def、await、asyncio.run()
* ⚠️ 常见误区：await 只对异步函数有效，普通函数（如 time.sleep）无法被挂起。在 async 函数里，禁止用 time.sleep，必须用 asyncio.sleep。

````python
import asyncio
import time

async def task(name: str, delay: int):
    print(f"任务 {name} 开始，等待 {delay}s")
    await asyncio.sleep(delay)  # 模拟 I/O 阻塞
    print(f"任务 {name} 完成")
    return f"{name} 结果"

async def main():
    # 并发执行多个任务（总耗时约 2s，而非 3s）
    results = await asyncio.gather(
        task("A", 2),
        task("B", 1),
        task("C", 1.5)
    )
    print(f"所有结果: {results}")

# 启动入口（脚本必须用此方法运行）
asyncio.run(main())
## asyncio.run() 不能在已有事件循环的环境中调用，而 Jupyter Notebook / IPython 本身就在后台运行着一个事件循环
````

## 6. datetime —— 时间处理
* 时间处理的核心是：当前时刻、格式化字符串、时间差（timedelta）
* ⚠️ 面试常问陷阱：datetime.now() 是本地时间（依赖系统时区），服务器部署时容易混淆。Web 开发中，存储数据库建议一律用 UTC 时间，显示时再转本地时区。

````python
from datetime import datetime, timedelta, timezone

# ---------- 获取当前时间 ----------
now = datetime.now()                          # 本地时间（无时区）
utc_now = datetime.now(timezone.utc)          # UTC 时间（推荐）晚8h
print(now)
print(utc_now)

# ---------- 格式化与解析 ----------
# 对象 → 字符串（给用户看）
formatted = now.strftime("%Y-%m-%d %H:%M:%S")  # 2026-06-23 14:30:00
print(formatted)
print(utc_now.strftime("%Y-%m-%d %H:%M:%S"))

# 字符串 → 对象（解析用户输入/API 时间）
time_str = "2025-01-01 12:00:00"
parsed = datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S")
print(parsed)

# ---------- 时间计算（timedelta） ----------
yesterday = now - timedelta(days=1)
next_hour = now + timedelta(hours=1)
print(yesterday.strftime("%Y-%m-%d"))

# ---------- 时间戳（秒数）互换 ----------
timestamp = now.timestamp()   # 转 Unix 时间戳
recovered = datetime.fromtimestamp(timestamp)
print(timestamp)
print(recovered)
````

## 综合实战：一条日志记录完整 API 请求
* 下面演示把上述库串起来：读取环境变量 → 发起异步请求 → 记录日志 → 解析 JSON → 计算耗时。

````python
import os
import json
import asyncio
import logging
from pathlib import Path
from datetime import datetime

# 1. 配置日志
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")
logger = logging.getLogger(__name__)

# 2. 读取环境变量（模拟 API Key）
API_KEY = os.getenv("MOCK_API_KEY", "default-key-123")

async def fetch_data():
    start = datetime.now()
    logger.info(f"开始请求，API Key: {API_KEY[:4]}***")
    
    # 模拟异步网络请求（实际用 aiohttp/httpx）
    await asyncio.sleep(1.5)
    mock_response = {"status": "ok", "data": [1, 2, 3]}
    
    # 3. 处理 JSON
    json_str = json.dumps(mock_response, indent=2)
    logger.info(f"收到响应：\n{json_str}")
    
    # 4. 计算耗时
    elapsed = (datetime.now() - start).total_seconds()
    logger.info(f"请求完成，耗时 {elapsed:.2f}s")
    return mock_response

if __name__ == "__main__":
    asyncio.run(fetch_data())
````
