```python
def count_words(text):
    wordDict = {}
    spliteData = text.split()
    print(spliteData)
    for word in spliteData:
        lowerV = word.lower()
        print(lowerV)
        if lowerV in wordDict:
            wordDict[lowerV] += 1
        else:
            wordDict[lowerV] = 1
    return wordDict

count_words("Hello hello world")
```

```python
# - 正常情况返回 `a / b` 的结果（保留两位小数）
# - `b` 为 0 时返回 `"错误：除数不能为零"`
# - 参数不是数字时返回 `"错误：参数必须是数字"`
# - 无论成功失败，最后都打印 `"计算完成

def safe_divide(a, b):
    try:
        if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
            return "错误：参数必须是数字"
        result = round(a / b, 2)
        return result
    except ZeroDivisionError:
        return "错误：除数不能为零"
    finally:
        print("计算完成")

safe_divide(1,3)
```

```python
headers = ["Name", "Age", "City"]
rows = [
    ["Alice", 25, "Beijing"],
    ["Bob", 30, "Shanghai"],
    ["Charlie", 22, "Guangzhou"],
]

def format_table(rows, headers):
    col_widths = []
    for i, h in enumerate(headers):
        max_w = len(str(h))
        for row in rows:
            max_w = max(max_w, len(str(row[i])))
        col_widths.append(max_w)
    print(col_widths)
    # t = [str(v).ljust(col_widths[i]) for i, v in enumerate(headers)] 列表会一次性生成所有结果
    def fmt_row(values):
        return " | ".join(str(v).ljust(col_widths[i]) for i, v in enumerate(values))

    lines = [fmt_row(headers)]
    lines.append("-" * len(lines[0]))
    for row in rows:
        lines.append(fmt_row(row))
    return "\n".join(lines)

print(format_table(rows, headers))

```

```python
# 编写函数 `retry(func, max_attempts=3, delay=0)`：

# - 调用 `func()`，如果成功则返回结果
# - 如果抛出异常，最多重试 `max_attempts` 次
# - 每次重试前等待 `delay` 秒（用 `time.sleep`）
# - 如果所有尝试都失败，抛出最后一次的异常
# - 每次重试时打印 `"第 N 次重试..."`

# 示例用法
import random
import time

def unstable_operation():
    if random.random() < 0.7:
        raise ValueError("操作失败")
    return "成功！"

def retry(func, max_attempts=3, delay=0):
    last_exception = None
    for attempt in range(1, max_attempts + 1):
        try:
            return func()
        except Exception as e:
            last_exception = e
            if attempt < max_attempts:
                print(f"第 {attempt} 次重试...")
                if delay:
                    time.sleep(delay)
    raise last_exception


result = retry(unstable_operation, max_attempts=5, delay=1)
print(result)
```

```python
# 编写函数 `process_csv_data(lines, **transforms)`：

# - `lines` 是一个字符串列表，第一行为表头，后续为数据行，逗号分隔
# - `**transforms` 是字段名到转换函数的映射
# - 返回处理后的字典列表

lines = [
    "name,age,score,active",
    "Alice,25,88.5,true",
    "Bob,30,92.0,false",
    "Charlie,22,76.3,true",
]

def to_int(x): return int(x)
def to_float(x): return float(x)
def to_bool(x): return x.lower() == "true"

# 返回 [
#     {"name": "Alice", "age": 25, "score": 88.5, "active": True},
#     {"name": "Bob", "age": 30, "score": 92.0, "active": False},
#     {"name": "Charlie", "age": 22, "score": 76.3, "active": True},
# ]

# **额外要求：** 如果某行字段数与表头不一致，跳过该行并打印警告。

def process_csv_data(lines, **transforms):
    headers = lines[0].split(",")
    result = []
    for i, line in enumerate(lines[1:], start=2):
        fields = line.split(",")
        if len(fields) != len(headers):
            print(f"警告：第 {i} 行字段数不匹配，已跳过")
            continue
        row = {}
        for h, v in zip(headers, fields):
            if h in transforms:
                row[h] = transforms[h](v.strip())
            else:
                row[h] = v.strip()
        result.append(row)
    return result

result = process_csv_data(lines, age=to_int, score=to_float, active=to_bool)

print(result)
```
