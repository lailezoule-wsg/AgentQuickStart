## dict.fromkeys()
* dict.fromkeys() 是 Python 字典的一个类方法，用于创建一个新字典，其键来自给定的可迭代对象，所有键的值都设为同一个初始值

````python
### 1. 语法与参数
dict.fromkeys(iterable, [value])
iterable（必填）：任何可迭代对象（列表、元组、集合、字符串等），用于提供字典的键。
value（可选）：所有键共同对应的值。如果不提供，默认是 None

### 2. 基本用法
# 创建一个空字典的键
keys = ['name', 'age', 'city']
d1 = dict.fromkeys(keys)
print(d1)  # 输出: {'name': None, 'age': None, 'city': None}

# 设置默认值
d2 = dict.fromkeys(keys, 0)
print(d2)  # 输出: {'name': 0, 'age': 0, 'city': 0}

# 使用字符串（每个字符成为键）
d3 = dict.fromkeys('abc', 1)
print(d3)  # 输出: {'a': 1, 'b': 1, 'c': 1}

# 使用集合（顺序不保证）
d4 = dict.fromkeys({1, 2, 3}, 'x')
print(d4)  # 输出: {1: 'x', 2: 'x', 3: 'x'}

### 3. ⚠️ 关键陷阱：所有键共享同一个值对象
"""
这是 fromkeys 最容易被误用的地方，也是面试高频考点。
如果 value 是可变对象（如列表、字典、集合），所有键会指向同一个对象，修改其中一个会影响到所有键！
"""
keys = ['a', 'b', 'c']
d = dict.fromkeys(keys, [])   # 所有键共享同一个空列表

print(d)  # {'a': [], 'b': [], 'c': []}

# 看似只给 'a' 追加了元素
d['a'].append(1)

print(d)  # 输出: {'a': [1], 'b': [1], 'c': [1]}  ❌ 全部被改了！

### 为什么会这样？
"""
fromkeys 只是把 [] 这个对象的引用复制了三份，并没有创建三个独立的新列表。
解决方法：使用字典推导式，为每个键单独创建新的可变对象。
"""
python
# ✅ 正确写法：每个键拥有独立的列表
d = {key: [] for key in ['a', 'b', 'c']}
d['a'].append(1)
print(d)  # 输出: {'a': [1], 'b': [], 'c': []}

### 4. 与 dict 构造函数的区别
"""
方法	                        说明
dict.fromkeys(keys, value)	    所有键共享同一个 value（引用同一对象）。
dict(zip(keys, values))	        键和值一一对应，可以指定不同的值。
{key: value for key in keys}	每个键都独立计算值（适合需要创建新可变对象时）。
"""
### 5. 实战应用场景
"""
初始化计数器（统计频率）：
words = ['apple', 'banana', 'apple', 'orange']
freq = dict.fromkeys(words, 0)
for word in words:
    freq[word] += 1
print(freq)  # {'apple': 2, 'banana': 1, 'orange': 1}

构建默认配置模版：
config_template = dict.fromkeys(['host', 'port', 'debug'], 'default')
# 然后根据环境覆盖部分值


快速将列表转换为字典（键为列表元素，值为固定占位符）。
"""
````

## def f(a, b, /, c, *, d):
/（位置仅限参数分隔符）：
在 / 之前的参数（这里是 a, b）只能通过位置传递，不能使用关键字形式。
调用时，1 赋给 a，2 赋给 b，完全符合规则。

*（关键字仅限参数分隔符）：
在 * 之后的参数（这里是 d）只能通过关键字传递，不能使用位置形式。
调用时，d=4 符合规则。

c（普通参数）：
位于 / 和 * 之间的参数，既可以按位置传递，也可以按关键字传递。
这里 3 被按位置赋给了 c，没有问题。

## 多元素排序： stu = sorted(students,key=lambda stu:(-stu["score"],stu["name"]))

````python
# 给定一个学生列表：
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78},
    {"name": "Diana", "score": 95},
]
# - 按分数从高到低排序
# - 分数相同时按名字字母序排序
# - 输出排序后的名字列表（用列表推导式提取）
# 期望输出: ["Diana", "Bob", "Alice", "Charlie"]

stu = sorted(students,key=lambda stu:(-stu["score"],stu["name"]))

print(stu)

stuName = [s["name"] for s in stu]
print(stuName)
````

## logging 日志写入

````python
import logging
def setup_logger(name, log_file, level=logging.INFO):
    logger = logging.getLogger(name)
    logger.setLevel(level)
    formatter = logging.Formatter("%(asctime)s - %(name)s - %(levelname)s - %(message)s")

    # FileHandler
    fh = logging.FileHandler(filename=log_file,encoding="utf-8")
    fh.setFormatter(formatter)

    # StreamHandler
    sh = logging.StreamHandler()
    sh.setFormatter(formatter)

    logger.addHandler(fh)
    logger.addHandler(sh)

    return logger


logger = setup_logger(__name__,"lf.log")
logger.info("程序启动")
logger.error("发生了一个错误")
````

## functools 提供了一组高阶函数（操作函数的函数），主要用来增强、优化或简化函数的行为。它是 Python 函数式编程的重要工具库


### 📦 最常用的 5 个工具

| 工具                  | 作用                                 | 经典场景                   |
| :-------------------- | :----------------------------------- | :------------------------- |
| **`@lru_cache`**      | 缓存函数结果，避免重复计算           | 递归、斐波那契、数据库查询 |
| **`@wraps`**          | 保留被装饰函数的元信息（名称、文档） | 自定义装饰器时必用         |
| **`partial`**         | 固定函数的部分参数，生成新函数       | 简化函数调用，减少重复传参 |
| **`reduce`**          | 对可迭代对象累计计算                 | 求积、拼接字符串、连续操作 |
| **`@singledispatch`** | 根据参数类型执行不同实现             | 泛型函数（类似函数重载）   |

### 📝 总结一句话

`functools` 是 Python 标准库中的 **“函数工具箱”**，用来：

- **提升性能**（缓存、记忆化）
- **简化代码**（固定参数、减少重复）
- **增强装饰器**（保留元信息）
- **实现泛型**（单分派函数）

在日常开发中，**`partial` 和 `@lru_cache` 是最常用的**，掌握它们可以写出更简洁、更高效的代码

````python
# 1. @lru_cache —— 自动缓存
from functools import lru_cache

@lru_cache(maxsize=128)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# 第一次计算会递归，后续相同参数直接从缓存取，速度极快
print(fibonacci(40))  # 102334155

# 2. @wraps —— 保留原函数信息
from functools import wraps

def log_call(func):
    @wraps(func)  # 不加这一行，help(hello) 会显示 inner
    def wrapper(*args, **kwargs):
        print(f"调用 {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@log_call
def hello():
    """这是 hello 函数的文档"""
    print("hello")

hello()
print(hello.__name__)  # hello（不加 wraps 会变成 wrapper）
print(hello.__doc__)   # 这是 hello 函数的文档（保留）

# 3. partial —— 固定参数
from functools import partial

def power(base, exponent):
    return base ** exponent

# 固定 exponent=2，生成平方函数
square = partial(power, exponent=2)
print(square(5))  # 25

# 固定 exponent=3，生成立方函数
cube = partial(power, exponent=3)
print(cube(3))   # 27

# 4. reduce —— 累计计算
from functools import reduce

nums = [1, 2, 3, 4, 5]
# 计算乘积
product = reduce(lambda x, y: x * y, nums)
print(product)  # 120

# 5. @singledispatch —— 函数重载
from functools import singledispatch

@singledispatch
def process(arg):
    print(f"默认: {arg}")

@process.register(int)
def _(arg):
    print(f"整数: {arg}")

@process.register(str)
def _(arg):
    print(f"字符串: {arg}")

process(10)    # 整数: 10
process("hi")  # 字符串: hi
process([1,2]) # 默认: [1, 2]
````

## @dataclass 魔术方法

给出每个魔术方法最经典的**使用示例**，让你一看就懂、拿来就能用。

------

#### 1. 算术运算符：`__add__`（实现向量加法）

python

```
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        # 要求 other 也是 Vector
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3)  # Vector(4, 6)
```



------

#### 2. 比较运算符：`__lt__`（实现按年龄排序）

python

```
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __lt__(self, other):
        return self.age < other.age

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"

people = [
    Person("Alice", 30),
    Person("Bob", 25),
    Person("Charlie", 35)
]
sorted_people = sorted(people)  # 按年龄升序
print(sorted_people)
# [Person('Bob', 25), Person('Alice', 30), Person('Charlie', 35)]
```



> 只要定义了 `__lt__`，`sorted()`、`min()`、`max()` 都能自动工作。

------

#### 3. 容器模拟：`__getitem__`（让对象像列表一样取值）

python

```
class MyList:
    def __init__(self, items):
        self.items = items

    def __getitem__(self, index):
        return self.items[index]

    def __len__(self):
        return len(self.items)

ml = MyList([10, 20, 30])
print(ml[0])    # 10
print(ml[1:3])  # [20, 30]（切片也支持！）
print(len(ml))  # 3
```



> 实现了 `__getitem__` 后，还可以**迭代**（`for item in ml:` 自动生效）。

#### 4. 上下文管理：`__enter__` / `__exit__`（自定义 with 语句）

python

```
class ManagedFile:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.file = None

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.file:
            self.file.close()

# 使用 with 自动管理文件关闭
with ManagedFile("test.txt", "w") as f:
    f.write("Hello, World!")
```



> 这就是 Python 中 `with open(...) as f` 的底层实现原理。

#### 5. 描述符协议：`__get__` / `__set__` / `__delete__`（实现属性校验）

描述符是 `@property` 的底层，通常用于**多个属性复用同一套校验逻辑**。

python

```
class PositiveNumber:
    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        return instance.__dict__.get(self.name, 0)

    def __set__(self, instance, value):
        if value < 0:
            raise ValueError(f"{self.name} 不能为负数")
        instance.__dict__[self.name] = value

class Order:
    price = PositiveNumber()
    quantity = PositiveNumber()

    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity

order = Order(100, 2)
print(order.price)   # 100
# order.price = -10  # 抛出 ValueError: price 不能为负数
```



> 描述符让多个属性共享同一套校验规则，代码复用性极强。

## asyncio.Queue 用法

* 先进先出（FIFO，First-In-First-Out）

* asyncio.Queue 是异步编程中的线程安全队列，专用于协程之间传递数据。它和 queue.Queue（同步队列）类似，但所有操作都是异步的（await），不会阻塞事件循环。

* 即使有多个生产者同时放入，多个消费者同时取出，队列本身仍然保证 FIFO 顺序——先 put 成功的元素，会被先 get 到。

* 这是构建生产者-消费者模式、流式处理、任务池的核心工具

| 队列类型                | 顺序                             | 适用场景                 |
| :---------------------- | :------------------------------- | :----------------------- |
| `asyncio.Queue`         | **FIFO（先进先出）**             | 默认，最常用（任务排队） |
| `asyncio.LifoQueue`     | **LIFO（后进先出）**             | 栈结构，如撤销操作       |
| `asyncio.PriorityQueue` | **优先级顺序**（数字越小越先出） | 紧急任务优先处理         |

````python
import asyncio

async def demo_order():
    q = asyncio.Queue()
    
    # 按顺序放入 1, 2, 3, 4, 5
    for i in range(1, 6):
        await q.put(i)
        print(f"✅ 放入: {i}")
    
    print("-" * 30)
    
    # 按顺序取出
    while not q.empty():
        item = await q.get()
        print(f"📦 取出: {item}")
        q.task_done()

asyncio.run(demo_order())

# 输出：
# ✅ 放入: 1
# ✅ 放入: 2
# ✅ 放入: 3
# ✅ 放入: 4
# ✅ 放入: 5
# ------------------------------
# 📦 取出: 1
# 📦 取出: 2
# 📦 取出: 3
# 📦 取出: 4
# 📦 取出: 5
````
