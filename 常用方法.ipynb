## dict.fromkeys()
* dict.fromkeys() 是 Python 字典的一个类方法，用于创建一个新字典，其键来自给定的可迭代对象，所有键的值都设为同一个初始值

```python
### 1. 语法与参数
dict.fromkeys(iterable, [value])
iterable（必填）：任何可迭代对象（列表、元组、集合、字符串等），用于提供字典的键。
value（可选）：所有键共同对应的值。如果不提供，默认是 None

### 2. 基本用法
# 创建一个空字典的键
keys = ['name', 'age', 'city']
d1 = dict.fromkeys(keys)
print(d1)  # 输出: {'name': None, 'age': None, 'city': None}

# 设置默认值
d2 = dict.fromkeys(keys, 0)
print(d2)  # 输出: {'name': 0, 'age': 0, 'city': 0}

# 使用字符串（每个字符成为键）
d3 = dict.fromkeys('abc', 1)
print(d3)  # 输出: {'a': 1, 'b': 1, 'c': 1}

# 使用集合（顺序不保证）
d4 = dict.fromkeys({1, 2, 3}, 'x')
print(d4)  # 输出: {1: 'x', 2: 'x', 3: 'x'}

### 3. ⚠️ 关键陷阱：所有键共享同一个值对象
"""
这是 fromkeys 最容易被误用的地方，也是面试高频考点。
如果 value 是可变对象（如列表、字典、集合），所有键会指向同一个对象，修改其中一个会影响到所有键！
"""
keys = ['a', 'b', 'c']
d = dict.fromkeys(keys, [])   # 所有键共享同一个空列表

print(d)  # {'a': [], 'b': [], 'c': []}

# 看似只给 'a' 追加了元素
d['a'].append(1)

print(d)  # 输出: {'a': [1], 'b': [1], 'c': [1]}  ❌ 全部被改了！

### 为什么会这样？
"""
fromkeys 只是把 [] 这个对象的引用复制了三份，并没有创建三个独立的新列表。
解决方法：使用字典推导式，为每个键单独创建新的可变对象。
"""
python
# ✅ 正确写法：每个键拥有独立的列表
d = {key: [] for key in ['a', 'b', 'c']}
d['a'].append(1)
print(d)  # 输出: {'a': [1], 'b': [], 'c': []}

### 4. 与 dict 构造函数的区别
"""
方法	                        说明
dict.fromkeys(keys, value)	    所有键共享同一个 value（引用同一对象）。
dict(zip(keys, values))	        键和值一一对应，可以指定不同的值。
{key: value for key in keys}	每个键都独立计算值（适合需要创建新可变对象时）。
"""
### 5. 实战应用场景
"""
初始化计数器（统计频率）：
words = ['apple', 'banana', 'apple', 'orange']
freq = dict.fromkeys(words, 0)
for word in words:
    freq[word] += 1
print(freq)  # {'apple': 2, 'banana': 1, 'orange': 1}

构建默认配置模版：
config_template = dict.fromkeys(['host', 'port', 'debug'], 'default')
# 然后根据环境覆盖部分值


快速将列表转换为字典（键为列表元素，值为固定占位符）。
"""
```

## def f(a, b, /, c, *, d):
/（位置仅限参数分隔符）：
在 / 之前的参数（这里是 a, b）只能通过位置传递，不能使用关键字形式。
调用时，1 赋给 a，2 赋给 b，完全符合规则。

*（关键字仅限参数分隔符）：
在 * 之后的参数（这里是 d）只能通过关键字传递，不能使用位置形式。
调用时，d=4 符合规则。

c（普通参数）：
位于 / 和 * 之间的参数，既可以按位置传递，也可以按关键字传递。
这里 3 被按位置赋给了 c，没有问题。

## zip() 是 Python 内置函数，核心作用是将多个可迭代对象（列表、元组、字符串等）“拉链式”地聚合在一起，按索引位置配对，生成一个由元组组成的迭代器
* 所有生成器都是迭代器，但迭代器不一定是生成器。大部分情况下，如果你不需要自定义迭代器的底层行为，直接用生成器就行了。

```python
# 给定一个二维矩阵：
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# 用列表推导式实现**矩阵转置**（行变列），输出：
# [[1, 4, 7], [2, 5, 8], [3, 6, 9]]
dt = [list(num) for num in zip(*matrix)]
```

## 多元素排序： stu = sorted(students,key=lambda stu:(-stu["score"],stu["name"]))

```python
# 给定一个学生列表：
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78},
    {"name": "Diana", "score": 95},
]
# - 按分数从高到低排序
# - 分数相同时按名字字母序排序
# - 输出排序后的名字列表（用列表推导式提取）
# 期望输出: ["Diana", "Bob", "Alice", "Charlie"]

stu = sorted(students,key=lambda stu:(-stu["score"],stu["name"]))

print(stu)

stuName = [s["name"] for s in stu]
print(stuName)
```

## all() 判断一个可迭代对象（iterable）里的所有元素，是否全部为真（Truthy）
* 可迭代对象：列表、集合、元组、生成器、range、空迭代器（all([])、all({})、all(())）、字典（key）

```python
any([False, False, True])  # True
all([True, True, False])  # False
```

## logging 日志写入

```python
import logging
def setup_logger(name, log_file, level=logging.INFO):
    logger = logging.getLogger(name)
    logger.setLevel(level)
    formatter = logging.Formatter("%(asctime)s - %(name)s - %(levelname)s - %(message)s")

    # FileHandler
    fh = logging.FileHandler(filename=log_file,encoding="utf-8")
    fh.setFormatter(formatter)

    # StreamHandler
    sh = logging.StreamHandler()
    sh.setFormatter(formatter)

    logger.addHandler(fh)
    logger.addHandler(sh)

    return logger


logger = setup_logger(__name__,"lf.log")
logger.info("程序启动")
logger.error("发生了一个错误")
```
